```python
import requests
import shutil

height=0.00445
width=0.00510260779

xmin=76.846585
ymax=28.866605-68*height
ymin=ymax-height
xmax=xmin+width

repeat=0
count=6733


res_x=500
res_y=500

for i in range(34):
    j=0
    while(j<99):
        image_url  = "http://server.arcgisonline.com/arcgis/rest/services/World_Imagery/MapServer/export?bbox="+str(xmin)+"%2C"+str(ymin)+"%2C"+str(xmax)+"%2C"+str(ymax)+"&bboxSR=4326&layers=&layerDefs=&size="+str(res_x)+"%2C"+str(res_y)+"&imageSR=&format=png&transparent=false&dpi=&time=&layerTimeOptions=&dynamicLayers=&gdbVersion=&mapScale=&rotation=&datumTransformations=&layerParameterValues=&mapRangeValues=&layerRangeValues=&f=image"
        resp = requests.get(image_url, stream=True)
        #print(image_url)

        if(resp.status_code==200):
            if(res_x == 500):
                local_file = open('img_'+str(count)+'.png', 'wb')
            else:
                local_file = open('img_'+str(count)+'_450.png', 'wb')
            
            resp.raw.decode_content = True
            shutil.copyfileobj(resp.raw, local_file)
            text_file=open('img_locations'+'.txt','a')
            text_file.write('img_'+str(count)+','+str(xmin)+","+str(ymin)+","+str(xmax)+","+str(ymax)+"\n")
            del resp
            repeat=0
            xmin=xmax
            xmax=xmax+width
            print('img_'+str(count)+" downloaded",res_x,res_y)
            count+=1
            res_x=500
            res_y=500
            j+=1
            
        else:
            if(resp.status_code==500):
                del resp
                res_x=450
                res_y=450 
                
                
            if(repeat<=5):
                repeat+=1

                continue;
            else:
                print("Unable to download img"+str(count)+ "\nxmin="+str(xmin)+"\nymin="+str(ymin)+"\nxmax="+str(xmax)+"\nymax="+str(ymax))
                error_file=open('img_errors'+'.txt','a')
                error_file.write('img_'+str(count)+','+str(xmin)+","+str(ymin)+","+str(xmax)+","+str(ymax)+"\n")
                repeat=0
                xmin=xmax
                xmax=xmax+width
                count+=1
                res_x=500
                res_y=500
                j+=1
    
    ymax=ymin
    ymin=ymin-height 
    xmin = 76.846585
    xmax=xmin+width
    print("changed row")
text_file.close()
```
